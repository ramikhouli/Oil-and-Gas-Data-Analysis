```python
import numpy as np
import pandas as pd

import warnings
warnings.filterwarnings('ignore')

import matplotlib.pyplot as plt
```

### Exercise 1: Relative Permeability Plots.

```python
sw = np.linspace(0,1,100)

krw = sw**3 ; kro = (1-sw)**3
```

We can also customize our plotting styles: 'default' , 'seaborn' etc.. 

```python
plt.style.use('default')
```

```python
#step 1: Plotting the first plot
plt.plot(sw, krw)

#step 2: Plotting the second plot
plt.plot(sw, kro)
```

```python
#Adding a GRID

#step 1: Plotting the first plot
plt.plot(sw, krw, lw=5)   # lw = line width

#step 2: Plotting the second plot
plt.plot(sw, kro, lw=7)    # lw = line width

#step 3: Plotting the Grid
plt.grid()
```

```python
#plt.plot(sw, krw, sw, kro, lw=5)   # lw = line width
```

Customizing our plot with more details.

```python
#Labels and legends
plt.plot(sw, krw, label='Krw')
plt.plot(sw, kro, label='Kro')

plt.grid()
plt.legend()
```

```python
#X-label, Y-label & Title
plt.plot(sw, krw, lw=3, label='krw')
plt.plot(sw, kro, lw=3, label='kro')

plt.grid()

#X-label, Y-label
plt.xlabel('sw(fraction)')
plt.ylabel('krw(fraction)')
plt.legend()

#title
plt.title('Rel-perm Plots')
```

Histogram & Barplot

```python
import pandas as pd

phi_sst = np.random.normal(loc=25, scale=10, size=100)
phi_lst = np.random.normal(10, 5, 100)
df = pd.DataFrame({'S':phi_sst, 'C':phi_lst})

plt.hist(df['S'], alpha=0.5, label='S');
plt.hist(df['C'], alpha=0.5, label='C');

plt.legend()

#df
```

## Seaborn: Good for Statistical plots

```python
import seaborn as sns
sns.kdeplot(df['C'], color='orange') ; sns.distplot(df['S'], color='green')
```

```python
#Box plot

sns.boxplot(df['S'], color='green')
sns.boxplot(df['C'], color='orange')

plt.xlabel('Porosities')
```
